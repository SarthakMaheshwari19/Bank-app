```python
class BankAccount:
    def __init__(self,account_number,Name,Phone,balance=1000,):
        self.account_number=account_number
        self.balance=balance
        self.Name=Name
        self.Phone=Phone
        self.transaction_history = []
        
    def display_transaction_history(self):
        if self.transaction_history:
            print("Transaction History:")
            for transaction in self.transaction_history:
                print(transaction)
        else:
            print("No transactions yet")
                  
    def display_balance(self):
        print(f"Name:{self.Name} ->Phone:{self.Phone}\nAccount Number: {self.account_number} ->Balance: {self.balance}")


class SavingsAccount(BankAccount):
    interest_rate=0.02
                  
    def add_transaction(self,transaction_type,amount):
        self.transaction_history.append(f"{transaction_type} ---> {amount}")

    def add_interest(self):
        interest_earned= self.balance*self.interest_rate
        self.balance +=interest_earned
        self.add_transaction("Interest", interest_earned)
        print(f"Interest added ->New balance: {self.balance}")
        
    def deposit(self, amount):
        if amount>0:
            self.balance+= amount
            self.add_transaction("Deposit", amount)
            print(f"Deposited {amount} -> New balance: {self.balance}")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if 0<amount<=self.balance:
            self.balance -=amount
            self.add_transaction("Withdrawal", amount)
            print(f"Withdrew {amount} -> New balance:{self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds")

    def display_balance(self):
        super().display_balance()
        
    def display_transaction_history(self):
        super().display_transaction_history()
```

```python

savings_account = SavingsAccount(account_number="PNB213",Name="abc",Phone="1234567890")
c=False
while(c==False):
    case=int(input("\npress 1 For deposit\npress 2 For withdraw\npress 3 For account balance and info\npress 4 For Transaction history\n      TO EXIT press 5\n "))
    if(case==1):
        money=int(input("Enter the amount to deposit:"))
        savings_account.deposit(money)
    if(case==2):
        money=int(input("Enter the amount to withdraw:"))
        savings_account.withdraw(money)
    if(case==3):
        print("Here are your Account details:\n")
        savings_account.display_balance()
    if(case==4):
        print("Here are your Transaction details:")
        savings_account.display_transaction_history()
    if(case==5):
        c=True
savings_account.add_interest()
```
